```python
import math
import numpy as np
```

```python
def func(x):
  return math.sin(3*x)+0.3*x*x
```

```python
def iteration(K=10,init=4.5,alpha=0.1):
  step = 1
  inits = init
  x_list = []
  y_list = []
  y_0 = func(inits)
  x_list.append(inits)
  y_list.append(y_0)
  for i in range(K):
    
    y_next_1 = func(inits + alpha*step)
    y_next_2 = func(inits - alpha*step)
    if y_next_1>y_next_2:
      y_next = y_next_2
      inits = inits - alpha*step
      x_list.append(inits)
      y_list.append(y_next)
    else:
      y_next = y_next_1
      inits = inits + alpha*step
      x_list.append(inits)
      y_list.append(y_next)
  return x_list,y_list
# iteration()
```

```python
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.1) 
plt.plot(x,[func(x=i)  for i in x],)

xx,yy = iteration()
plt.plot(xx,yy,"rs-")
```

## ex2.3

```python
def func2(x1, x2):
  return np.tanh(4*x1+4*x2)+1+np.maximum(0.4*x1**2,1)
```

```python
def iteration(K=3,P=1000,init=np.array([2,2]).T,alpha=1):
  step = 1
  inits = init
  x_list = []
  y_list = []
  z_list = []
  y_0 = func2(inits[0],inits[1])
  y_next_temp = y_0
  x_list.append(inits[0])
  y_list.append(inits[1])
  z_list.append(y_0)
  for i in range(K):
    y_next = y_next_temp
    y_next_up = func2(inits[0],inits[1] + alpha*step)
    y_next_down = func2(inits[0],inits[1] - alpha*step)
    y_next_right = func2(inits[0] + alpha*step,inits[1])
    y_next_left = func2(inits[0] - alpha*step,inits[1])
    
    
    y_next_temp = min([y_next,y_next_up,y_next_left,y_next_right,y_next_down,])
    y_pos = np.argmin([y_next,y_next_up,y_next_left,y_next_right,y_next_down,])
    if y_pos == 0:
      inits[0],inits[1] = inits[0],inits[1]
    elif y_pos == 1:
      inits[0],inits[1] = inits[0],inits[1] + alpha*step
    elif y_pos == 2:
      inits[0],inits[1] = inits[0],inits[1] - alpha*step
    elif y_pos == 3:
      inits[0],inits[1] = inits[0] + alpha*step,inits[1] 
    elif y_pos == 4:
      inits[0],inits[1] = inits[0] - alpha*step,inits[1] 


    x_list.append(inits[0])
    y_list.append(inits[1])
    z_list.append(y_next_temp)
  return x_list,y_list,z_list
iteration()
```

```python

fig = plt.figure() #定义新的三维坐标轴
ax3 = plt.axes(projection='3d')
#定义三维数据
xx = np.arange(-6,6,0.05)
yy = np.arange(-6,6,0.05)
X, Y = np.meshgrid(xx, yy)
Z = func2(x1=X, x2=Y)
#作图
ax3.plot_surface(X,Y,Z,cmap='rainbow')
XX, YY, ZZ = iteration()
ax3.scatter3D(XX, YY, ZZ, cmap='Dark2')
XX, YY, ZZ
```

```python
fig = plt.figure() #定义新的三维坐标轴
ax3 = plt.axes(projection='3d')
#定义三维数据
xx = np.arange(-6,6,0.05)
yy = np.arange(-6,6,0.05)
X, Y = np.meshgrid(xx, yy)
Z = func3(x1=X, x2=Y)
#作图
ax3.plot_surface(X,Y,Z,cmap='rainbow')
zz = [0,0,0]
XX, YY, ZZ = iteration()
while abs(ZZ[-1]-zz[-1])>0.01:
  XX, YY, zz = iteration()
  
ax3.scatter3D(XX, YY, zz, cmap='Dark2')
# ax3.view_init(elev=120., azim=1)
```
